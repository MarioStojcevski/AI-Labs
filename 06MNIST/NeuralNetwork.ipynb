```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.special import expit
```

```python
class NeuralNetwork:
    
    def __init__(self, inodes, hnodes, onodes, lrate):
        
        self.inodes = inodes
        self.hnodes = hnodes
        self.onodes = onodes
        
        self.lrate = lrate
        
        self.wih = np.random.normal(0.0, pow(self.hnodes, -0.5), (self.hnodes, self.inodes))
        self.who = np.random.normal(0.0, pow(self.onodes, -0.5), (self.onodes, self.hnodes))
        
        self.activation_function = lambda x: expit(x)
        
    def train(self, inputs_list, targets_list):
        
        inputs = np.array(inputs_list, ndmin=2).T
        targets = np.array(targets_list, ndmin=2).T
        
        hidden_inputs = np.dot(self.wih, inputs)
        hidden_outputs = self.activation_function(hidden_inputs)
        
        final_inputs = np.dot(self.who, hidden_outputs)
        final_outputs = self.activation_function(final_inputs)
        
        # matrix arithmetic
        output_errors = targets - final_outputs
        hidden_errors = np.dot(self.who.T, output_errors)
        
        self.who += self.lrate * np.dot((output_errors * final_outputs * (1.0 - final_outputs)),
                                       np.transpose(hidden_outputs))
        self.wih += self.lrate * np.dot((hidden_errors * hidden_outputs * (1.0 - hidden_outputs)),
                                       np.transpose(inputs))
        
        pass
        
    def query(self, inputs_list):
        
        inputs = np.array(inputs_list, ndmin=2).T
        
        hidden_inputs = np.dot(self.wih, inputs)
        hidden_outputs = self.activation_function(hidden_inputs)
        
        final_inputs = np.dot(self.who, hidden_outputs)
        final_outputs = self.activation_function(final_inputs)
        
        return final_outputs
        
```

```python
learning_rate = 0.3

input_nodes = 784
hidden_nodes = 100
output_nodes = 10

network = NeuralNetwork(input_nodes, hidden_nodes, output_nodes, learning_rate)
```

```python
mnist = pd.read_csv('assignment5.xls')
mnist.head()
# training set 70%
# testing set  20%
# validate set 10%


full = len(mnist)
train_len = int(full * 0.7)
test_len = int(full * 0.2)
validate_len = int(full * 0.1)
print(f'{full} The lengths are: \nTRAIN \t- {train_len} \nTEST \t- {test_len} \nVAL \t- {validate_len}\n')

# get a copy of mnist and get training set
X = mnist.copy()
print(len(X))
train_set = X.sample(n = train_len)

# drop the testing set from X and get validation set
X = X.drop(train_set.index)
print(len(X))
test_set = X.sample(n = test_len)

# drop the validation set and now testing set is the rest
X = X.drop(test_set.index)
print(len(X))
validate_set = X

print()
print(f'Confirm lengths: {len(train_set)}, {len(test_set)}, {len(validate_set)}')
```
